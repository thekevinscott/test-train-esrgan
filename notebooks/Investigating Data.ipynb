```python
from IPython.display import display, HTML

ROOT = pathlib.Path('/notebooks')
VOLUMES = ROOT / 'code/volumes'

img_path = './sample_images/fundog.jpg'
def show_imageio(im):
    return Image.fromarray((im * 255).astype(np.uint8))
    


def check_folder(folder):
    OUTPUT = VOLUMES / folder / 'output'
    files = {}
    files_found = [f for f in os.listdir(OUTPUT) if f.endswith('.png')]
    if len(files_found):
        for file in files_found:
            parts = file.split('-')
            kind = parts[0]
            i = parts[1]
            version = parts[2].split('.')[0]
            files[kind] = files.get(kind, {})
            files[kind][i] = files.get(kind).get(i, {})
            files[kind][i][version] = file
        for kind in ['hr', 'lr']:
            images = files.get(kind)
            for i in sorted(images.keys()):
                image = images.get(i)
                for key in sorted(image.keys()):
                    file = image.get(key)
                    img_path = str(OUTPUT / file)
                    im = imageio.imread(img_path) / 255.0
                    print(file)
                    display(show_imageio(im))
    else:
        print('No files found')


check_folder('fc9199d1')
```
